각 일자별 확진자 비율 확인 

```
# 확진자 비율 확인 함수 구현

mypath = '/content/drive/My Drive/covid_audio_data/csv/'

filenames = os.listdir(mypath)
print(filenames)

for file in filenames:
    path = mypath + file
    print(path)
    covid = pd.read_csv(path)   
    ax = covid['covid_status'].value_counts().plot.bar( color=['blue', 'red', 'green', 'cyan', 'purple', 'pink'])
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    plt.show()
```
